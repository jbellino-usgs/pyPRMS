# **pyPRMS**
## DataFile Example
This example notebook demonstrates how to read a PRMS CBH and station data files using the `pyPRMS.datfile.DataFile` class to view and manipulate the data. The `data` attribute of the `DataFile` class returns a `dict` with variable names as the keys and `pandas.DataFrame` objects as the values. The `variables` attrbute returns a `dict` with variable names as the keys and number of data columns as the values.

```python
import os
import matplotlib.pyplot as plt
from pyPRMS.datfile import DataFile

datpth = os.path.join('..', 'data')
```

## Read a CBH Data File

```python
fname = os.path.join(datpth, 'puerto_rico', 'input', 'daymet_tmax_1980_2017.data')   
datafile = DataFile(fname)
print(datafile.variables)
datafile.data['tmax'].head()
```

## Plot CBH Data
Plot maximum tempterature for selected HRUs for the period 1995–1996.

```python
fig = plt.figure(figsize=(12, 5))

ax = fig.add_subplot(111)

varname = 'tmax'
ax.set_title(varname)

fname = os.path.join(datpth, 'puerto_rico', 'input', 'daymet_tmax_1980_2017.data')   
datafile = DataFile(fname)
data = datafile.data[varname]

hrulst = [77, 127, 489]
data['1995':'1996'][hrulst].plot(ax=ax)

plt.show()
```

## Read and Plot a Station Data File
